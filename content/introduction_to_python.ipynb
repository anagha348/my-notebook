## 🧠 Data types in python
A **data type** defines the type of value a variable can hold and determines what operations can be performed on it.

In Python, data types are broadly categorised into:

- **Scalar (basic) data types**
- **Non-scalar (collection) data types**

## 🔹 Scalar Data Types

Scalar data types represent a single value.

| Data Type | Meaning | Example |
|----------|--------|--------|
| `int` | Integer numbers (whole numbers) | 10, -5 |
| `float` | Decimal numbers | 3.14, -0.5 |
| `complex` | Numbers with real + imaginary part | 2 + 3j |
| `bool` | Boolean values (True/False) | True |
| `str` | Sequence of characters (text) | "Hello" |
| `NoneType` | Represents absence of value | None |

#### 🔹 Key Notes

- **String (`str`)**
  - Used for text data
  - Immutable (cannot be changed after creation)

- **Boolean (`bool`)**
  - Used in conditional statements and logical operations

- **None (`NoneType`)**
  - Represents no value or null
  - Commonly used to initialise variables

#### 🔹 Type Checking

We can check the data type of a variable using:

```python
type(variable)

````python
# Integer
age = 25
print("age =", age, "| type =", type(age))

# Float
temperature = 23.5
print("temperature =", temperature, "| type =", type(temperature))

# Complex
z = 2 + 3j
print("z =", z, "| type =", type(z))

# Boolean
is_valid = True
print("is_valid =", is_valid, "| type =", type(is_valid))

# String
name = "Python"
print("name =", name, "| type =", type(name))

# NoneType
value = None
print("value =", value, "| type =", type(value))
````

#### 🔹 Type Conversion

We can convert the data type of a variable into another type using:

```python
int(), float(), str(), bool()

````python
# Integer to Float
x = 10
y = float(x)
print("x:", x, "| type:", type(x))
print("y:", y, "| type:", type(y))

# Float to Integer
a = 3.7
b = int(a)
print("a:", a, "| type:", type(a))
print("b:", b, "| type:", type(b))  # Note: decimal part is removed

# String to Integer
s = "25"
num = int(s)
print("s:", s, "| type:", type(s))
print("num:", num, "| type:", type(num))

# String to Float
s2 = "3.5"
f = float(s2)
print("s2:", s2, "| type:", type(s2))
print("f:", f, "| type:", type(f))

# Important Case
# This will raise an error
# int("3.5")  ❌ ValueError
````

#### 🔹 Operations on Data Types

Each data type in Python supports specific operations. The type of data determines what operations are valid.

---

### 🔸 Numeric Types (`int`, `float`, `complex`)

These support **arithmetic operations**:

````python
a = 10
b = 3

print('a + b = ',a + b)   # Addition
print('a - b = ',a - b)   # Subtraction
print('a * b = ',a * b)   # Multiplication
print("a/b = ",a / b)   # Division (float result)
print("a//b = ",a // b)  # Floor division
print("a % b = ",a % b)   # Modulus (remainder)
print("a ** b = ",a ** b)  # Exponentiation
````

### 🔸 Boolean Types 

These support **logical operations**:

````python
x = True
y = False

print("x and y =", x and y)  # Logical AND
print("x or y =", x or y)   # Logical OR
print("not x =", not x)    # Logical NOT
````

### 🔸 String Types

These support **text operations**:

````python
s1 = "Hello"
s2 = "World"

print(s1 + " " + s2)   # Concatenation
print(s1 * 3)          # Repetition
print(s1.lower())      # Convert to lowercase
print(len(s1))         # Length of string
````

Find more in [https://docs.python.org/3/library/string.html](https://docs.python.org/3/library/string.html)


### 🔸 Comparison Operations (All Comparable Types)

Used to compare values → result is always True or False:

````python
a = 5
b = 10

print("a == b =",a==b)  # Equal
print("a != b =", a!=b)  # Not equal
print("a < b =", a<b)   # Less than
print("a > b =",a<b)   # Greater than
````

### ⚠️ Invalid Operations

Not all operations are allowed between different data types.  
Operations must be **type-compatible**, meaning Python should be able to interpret how to combine the values.

---

#### 🔸 Example of Invalid Operation

```python
10 + "5"   # ❌ TypeError

### 🔹 Multiline Strings and f-Strings

---

#### 🔸 Multiline Strings

Multiline strings allow you to write text across multiple lines using triple quotes (`'''` or `"""`).


````python
text = """This is a multiline string.
It spans across multiple lines.
Useful for documentation or long text."""
print(text)
````

The \n escape character is used to create a new line within a string.

````python
text = "This is line one.\nThis is line two.\nThis is line three."
print(text)
````

f-strings allow embedding variables or expressions directly inside strings using {}.

🔍 Why use f-strings?
- Cleaner and more readable than concatenation
- Supports expressions inside {}

````python
x = 10
y = 5

print(f"Sum of {x} and {y}: {x + y}")
print(f"Square of {x}: {x**2}")
````

````python
name = "Alice"
age = 25

print(f"My name is {name} and I am {age} years old.")
````

## 🧠 Data Structures in Python

Data structures are used to store and organise multiple values in a single variable.

Unlike scalar data types (which store a single value), data structures allow us to work with collections of data efficiently.

---

### 🔹 Types of Data Structures in Python

| Data Structure | Description | Example |
|---------------|------------|--------|
| `list` | Ordered, mutable collection | [1, 2, 3] |
| `tuple` | Ordered, immutable collection | (1, 2, 3) |
| `set` | Unordered, unique elements | {1, 2, 3} |
| `dict` | Key-value pairs | {"a": 1, "b": 2} |

---

### 🔍 Key Properties

| Property     | List | Tuple | Set | Dictionary |
|-------------|------|-------|-----|-----------|
| Ordered     | Yes   | Yes    | No  | Yes (Python 3.7+) |
| Mutable     | Yes   | No    | Yes  | Yes |
| Allows Duplicates | Yes | Yes | No | Keys - No, Values - Yes |

---

### 🔹 Why Use Data Structures?

- Store multiple values efficiently  
- Perform operations like search, insert, delete  
- Organise data for analysis and algorithms  

````python
# List
numbers = [1, 2, 3, 4]

# Tuple
coordinates = (10, 20)

# Set
unique_values = {1, 2, 3}

# Dictionary
student = {"name": "Alice", "age": 25}

print(numbers)
print(coordinates)
print(unique_values)
print(student)
````

## 🔹 Lists in Python

A **list** is an ordered and mutable collection of elements.

- Defined using square brackets `[]`
- Can store elements of different data types


````python
numbers = [1, 2, 3, 4, 5]
mixed = [1, "hello", 3.5, True]

print(numbers)
print(mixed)
````

### 🔸 Accessing Elements (Indexing)

Lists are **indexed**, starting from `0`.

- The first element → index `0`  
- The second element → index `1`  
- Negative indexing allows access from the end  
  - `-1` → last element  
  - `-2` → second last element  

````python
numbers = [10, 20, 30, 40]

print(numbers[0])   # First element → 10
print(numbers[2])   # Third element → 30
print(numbers[-1])  # Last element → 40
````

### 🔸 Slicing

Slicing allows accessing a range of elements from a list.

**Syntax:**
```python
list[start : end : step]

````python
numbers = [10, 20, 30, 40, 50]

print(numbers[1:4])   # [20, 30, 40]
print(numbers[2:])    # [30, 40, 50]
print(numbers[:3])    # [10, 20, 30]
print(numbers[::2])   # [10, 30, 50]
print(numbers[::-1])  # [50, 40, 30, 20, 10] (reverse)
````

🔸 List is Mutable

Lists can be modified after creation.

````python
numbers = [1, 2, 3]
print(numbers)
numbers[0] = 10   # Modify element
print(numbers)

numbers.append(4)  # Add element
print(numbers)
````

🔸 Common List Methods

````python
nums = [3, 1, 4]

nums.append(5)      # Add element at end → [3, 1, 4, 5]
nums.insert(1, 2)   # Insert at index 1 → [3, 2, 1, 4, 5]
nums.remove(4)      # Remove value 4 → [3, 2, 1, 5]
nums.pop()          # Remove last element → [3, 2, 1]
nums.sort()         # Sort list → [1, 2, 3]

print(nums)
````

🔸 Iterating through a List using **For Loop**

````python
numbers = [1, 2, 3]

for num in numbers:
    print(num)
````

````python
number = [1,2,3]

for i in range(len(number)):
    print(f"{numbers[i]} is at index {i}")
````

````python
numbers = [100, 200, 300]

for index, value in enumerate(numbers):
    print(index, value)
````

## 🔹 Tuples in Python

A **tuple** is an ordered and immutable collection of elements.

- Defined using parentheses `()`
- Similar to lists, but **cannot be modified after creation**

---

### 🔸 Creating a Tuple

````python
numbers = (1, 2, 3, 4)
mixed = (1, "hello", 3.5)

print(numbers)
print(mixed)
````

🔸 Accessing Elements (Indexing)

Tuples support indexing just like lists.

````python
numbers = (10, 20, 30, 40)

print(numbers[0])   # First element
print(numbers[-1])  # Last element
````

🔸 Slicing

````python
numbers = (10, 20, 30, 40, 50)

print(numbers[1:4])   # (20, 30, 40)
````

🔸 Tuple is Immutable

````python
numbers = (1, 2, 3)

# numbers[0] = 10   ❌ TypeError
````

🔸 Useful Tuple Methods

````python
nums = (1, 2, 2, 2, 3, 2, 3)

print(nums.count(2))   # Count occurrences
print(nums.index(1))   # Find index
````

## 🔹 Sets in Python

A **set** is an unordered collection of unique elements.

- Defined using curly braces `{}`  
- Does **not allow duplicate values**  
- Elements are **not indexed**

---

### 🔸 Creating a Set


````python
numbers = {1, 2, 3, 4}
mixed = {1, "hello", 3.5}

print(numbers)
print(mixed)
````

🔸 Uniqueness (No Duplicates)

````python
nums = {1, 2, 2, 3, 3, 4}
print(nums)   # {1, 2, 3, 4}
````

🔸 Adding and Removing Elements

````python
nums = {1, 2, 3}

nums.add(4)        # Add element -> {1,2,3,4}
nums.remove(2)     # Remove element (error if not found) -> {1,3,4}
nums.discard(10)   # Safe remove (no error) -> {1,3,4}

print(nums)
````

🔸 Set Operations

Sets support mathematical operations like Union, Intersection...

````python
a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)   # Union → {1, 2, 3, 4, 5}
print(a & b)   # Intersection → {3}
print(a - b)   # Difference → {1, 2}
````

🔸 Iterating Through a Set

````python
nums = {1, 2, 3}

for num in nums:
    print(num)
````

🔸No indexing - unordered

````python
nums = {1, 2, 3}

# print(nums[0])   ❌ TypeError (sets are not indexed)
````

## 🔹 Dictionaries in Python

A **dictionary** is a collection of key–value pairs.

- Defined using curly braces `{}`  
- Each key maps to a value → `key: value`  
- Keys must be **unique and immutable**  

---

### 🔸 Creating a Dictionary

````python
student = {
    "name": "Alice",
    "age": 25,
    "course": "AI"
}

print(student)
````

🔸 Accessing Values

````python
student = {"name": "Alice", "age": 25}

print(student["name"])     # Access using key
print(student.get("age"))  # Safe access - gives no error if there is no such key
````

🔸 Adding and Updating Elements

````python
student = {"name": "Alice", "age": 25}

student["age"] = 26          # Update value
student["city"] = "Berlin"   # Add new key-value pair

print(student)
````

🔸 Removing Elements

````python
student = {"name": "Alice", "age": 25, "city": "Berlin"}

student.pop("age")     # Remove by key
del student["city"]    # Delete key-value pair

print(student)
````

🔸 Iterating Through a Dictionary

````python
student = {"name": "Alice", "age": 25}

for key in student:
    print(key, student[key])

# Recommended
for key, value in student.items():
    print(key, value)
````

🔸 Useful Dictionary Methods

````python
student = {"name": "Alice", "age": 25}

print(student.keys())    # All keys
print(student.values())  # All values
print(student.items())   # Key-value pairs
````
