### 5. Comparisons

### 6. Logical Operations

Logical operators combine boolean expressions.

```python
# Logical operators: and, or, not
age = 25
has_license = True
has_car = False

# AND - both must be True
can_drive = age >= 18 and has_license
print(f"Can drive: {can_drive}")

# OR - at least one must be True
can_travel = has_car or has_license
print(f"Can travel: {can_travel}")

# NOT - inverts the boolean
is_minor = not (age >= 18)
print(f"Is minor: {is_minor}")
```

```python
# Complex logical expressions
score = 85
attendance = 0.9

# Combine multiple conditions
passes_course = (score >= 60 and attendance >= 0.75) or score >= 90
print(f"Passes course: {passes_course}")

# Chaining comparisons
x = 15
in_range = 10 < x < 20  # Same as: 10 < x and x < 20
print(f"15 is between 10 and 20: {in_range}")
```

## Lists vs Tuples vs Sets vs Dicts

Each data structure has different performance characteristics and use cases:

| Structure | Ordered | Mutable | Duplicates | Use Case |
|-----------|---------|---------|------------|----------|
| `list`    | Yes     | Yes     | Yes        | Ordered collection, need to modify |
| `tuple`   | Yes     | No      | Yes        | Immutable sequence, dict keys, function returns |
| `set`     | No      | Yes     | No         | Fast membership testing, removing duplicates |
| `dict`    | Yes*    | Yes     | Keys: No   | Key-value mapping, fast lookup by key |

*Dicts maintain insertion order since Python 3.7

```python
# Lists - ordered, mutable, allows duplicates
fruits = ["apple", "banana", "cherry", "apple"]
fruits.append("date")
fruits[0] = "apricot"
print(f"List: {fruits}")
print(f"Length: {len(fruits)}, 'banana' at index: {fruits.index('banana')}")
```

```python
# Tuples - ordered, immutable, allows duplicates
coordinates = (10.5, 20.3, 30.1)
point = (1, 2)

# Tuple unpacking - very common in Python
x, y = point
print(f"Point: x={x}, y={y}")

# Tuples can be dict keys (lists cannot)
grid = {(0, 0): "origin", (1, 0): "right", (0, 1): "up"}
print(f"Grid lookup: {grid[(0, 0)]}")
```

```python
# Sets - unordered, mutable, no duplicates
unique_letters = {"a", "b", "c", "a", "b"}  # duplicates removed
print(f"Set: {unique_letters}")

# Set operations
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

print(f"Union: {set_a | set_b}")
print(f"Intersection: {set_a & set_b}")
print(f"Difference (a-b): {set_a - set_b}")
print(f"Symmetric difference: {set_a ^ set_b}")
```

```python
# Dicts - key-value pairs, ordered (3.7+), mutable
person = {"name": "Alice", "age": 30, "city": "Berlin"}

print(f"Name: {person['name']}")
print(f"Keys: {list(person.keys())}")
print(f"Values: {list(person.values())}")
print(f"Items: {list(person.items())}")
```

## Dictionary Methods: `get()`, `setdefault()`, and `Counter`

These methods make dictionary manipulation cleaner and safer.

```python
# dict.get() - safe access with default value
config = {"debug": True, "port": 8080}

# Without get() - raises KeyError if key doesn't exist
# timeout = config['timeout']  # KeyError!

# With get() - returns None or specified default
timeout = config.get("timeout")  # Returns None
timeout = config.get("timeout", 30)  # Returns 30
print(f"Timeout: {timeout}")

# get() doesn't modify the dict
print(f"Config unchanged: {config}")
```

```python
# dict.setdefault() - get with auto-insert
# Perfect for building lists inside dicts

# Grouping words by first letter
words = ["apple", "banana", "apricot", "blueberry", "cherry", "avocado"]
grouped = {}

for word in words:
    first_letter = word[0]
    # setdefault creates empty list if key doesn't exist, then appends
    grouped.setdefault(first_letter, []).append(word)

print(f"Grouped: {grouped}")
```

```python
# Dict comprehension - concise dictionary creation
squares = {x: x**2 for x in range(1, 6)}
print(f"Squares: {squares}")

# Filtering with dict comprehension
even_squares = {x: x**2 for x in range(1, 11) if x % 2 == 0}
print(f"Even squares: {even_squares}")

# Inverting a dictionary
original = {"a": 1, "b": 2, "c": 3}
inverted = {v: k for k, v in original.items()}
print(f"Inverted: {inverted}")
```

```python
# Counter from collections - counting made easy
text = "data science is the science of data"
word_counts = Counter(text.split())
print(f"Word counts: {word_counts}")
print(f"Most common 2: {word_counts.most_common(2)}")

# Counter arithmetic
c1 = Counter(["a", "b", "a"])
c2 = Counter(["a", "c"])
print(f"Counter sum: {c1 + c2}")
print(f"Counter subtract: {c1 - c2}")
```

## Sorting with `key=`

The `key` parameter lets you customize sorting behavior without modifying the original data.

```python
# sorted() returns new list, .sort() modifies in place
numbers = [3, 1, 4, 1, 5, 9, 2, 6]

# sorted() - creates new list
sorted_nums = sorted(numbers)
print(f"Original: {numbers}")
print(f"Sorted (new): {sorted_nums}")

# .sort() - modifies in place, returns None
numbers_copy = numbers.copy()
result = numbers_copy.sort()
print(f"After .sort(): {numbers_copy}")
print(f".sort() returns: {result}")  # None!
```

```python
# Sorting with key= for custom ordering
words = ["banana", "pie", "Washington", "a", "Book"]

# Sort by length
by_length = sorted(words, key=len)
print(f"By length: {by_length}")

# Sort case-insensitive
case_insensitive = sorted(words, key=str.lower)
print(f"Case insensitive: {case_insensitive}")

# Sort by last character
by_last = sorted(words, key=lambda x: x[-1])
print(f"By last char: {by_last}")
```

```python
# Sorting complex data structures
students = [
    {"name": "Alice", "grade": 85, "age": 22},
    {"name": "Bob", "grade": 90, "age": 20},
    {"name": "Charlie", "grade": 85, "age": 21},
]

# Sort by grade (descending)
by_grade = sorted(students, key=lambda s: s["grade"], reverse=True)
print("By grade (desc):")
for s in by_grade:
    print(f"  {s['name']}: {s['grade']}")

# Sort by multiple fields: grade (desc), then age (asc)
by_grade_age = sorted(students, key=lambda s: (-s["grade"], s["age"]))
print("\nBy grade desc, age asc:")
for s in by_grade_age:
    print(f"  {s['name']}: grade={s['grade']}, age={s['age']}")
```

## Nested Data Structures

Real-world data is often nested—lists of dicts, dicts of lists, etc. This is exactly what you get from JSON APIs.

```python
# Simulating JSON API response - list of dicts
api_response = [
    {
        "id": 1,
        "user": {"name": "Alice", "email": "alice@example.com"},
        "orders": [{"item": "laptop", "price": 999}, {"item": "mouse", "price": 29}],
    },
    {
        "id": 2,
        "user": {"name": "Bob", "email": "bob@example.com"},
        "orders": [{"item": "keyboard", "price": 79}],
    },
    {
        "id": 3,
        "user": {"name": "Charlie", "email": "charlie@example.com"},
        "orders": [],
    },
]

# Extracting nested data
print("All user names:")
names = [record["user"]["name"] for record in api_response]
print(names)
```

```python
# Calculate total order value per user
for record in api_response:
    user_name = record["user"]["name"]
    total = sum(order["price"] for order in record["orders"])
    print(f"{user_name}: ${total}")
```

```python
# Flatten nested structure: extract all orders with user info
flat_orders = []
for record in api_response:
    user_name = record["user"]["name"]
    for order in record["orders"]:
        flat_orders.append(
            {"user": user_name, "item": order["item"], "price": order["price"]}
        )

print("Flattened orders:")
for order in flat_orders:
    print(f"  {order}")
```

> **Gotcha: Safe nested access**  
> When accessing deeply nested structures, use `.get()` at each level or handle `KeyError` to avoid crashes when data is missing.

```python
# Safe nested access pattern
def safe_get(data, *keys, default=None):
    """Safely navigate nested dicts/lists."""
    for key in keys:
        try:
            data = data[key]
        except (KeyError, IndexError, TypeError):
            return default
    return data


# Test with missing data
incomplete_record = {"id": 4, "user": {"name": "Dave"}}
email = safe_get(incomplete_record, "user", "email", default="no email")
first_order = safe_get(incomplete_record, "orders", 0, "item", default="no orders")
print(f"Email: {email}")
print(f"First order: {first_order}")
```

## Performance Comparison: List vs Set Lookup

This visualization demonstrates why choosing the right data structure matters. Set lookup is O(1) while list lookup is O(n).

```python
# Benchmark list vs set membership testing
sizes = [100, 1000, 10000, 100000]
list_times = []
set_times = []

for size in sizes:
    # Create test data
    test_list = list(range(size))
    test_set = set(range(size))
    search_value = size - 1  # Worst case for list (last element)

    # Time list lookup
    list_time = timeit.timeit(lambda: search_value in test_list, number=1000)
    list_times.append(list_time)

    # Time set lookup
    set_time = timeit.timeit(lambda: search_value in test_set, number=1000)
    set_times.append(set_time)

# Visualization
fig, ax = plt.subplots(figsize=(10, 6))
x = range(len(sizes))
width = 0.35

bars1 = ax.bar(
    [i - width / 2 for i in x], list_times, width, label="List", color="#e74c3c"
)
bars2 = ax.bar(
    [i + width / 2 for i in x], set_times, width, label="Set", color="#2ecc71"
)

ax.set_xlabel("Collection Size")
ax.set_ylabel("Time (seconds) for 1000 lookups")
ax.set_title("List vs Set: Membership Testing Performance\n(Lower is Better)")
ax.set_xticks(x)
ax.set_xticklabels([f"{s:,}" for s in sizes])
ax.legend()
ax.set_yscale("log")  # Log scale to show both

# Add value labels
for bar, time in zip(bars1, list_times):
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f"{time:.4f}s",
        ha="center",
        va="bottom",
        fontsize=8,
    )
for bar, time in zip(bars2, set_times):
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f"{time:.4f}s",
        ha="center",
        va="bottom",
        fontsize=8,
    )

plt.tight_layout()
plt.show()

print(
    f"\nAt size 100,000: Set is {list_times[-1]/set_times[-1]:.0f}x faster than List!"
)
```

### Exercise 4: Flatten and Aggregate Nested Data

Given the nested sales data below, calculate:
1. Total revenue per region
2. The region with the highest total revenue

```python
sales_data = [
    {
        "region": "North",
        "sales": [{"product": "A", "amount": 100}, {"product": "B", "amount": 200}],
    },
    {
        "region": "South",
        "sales": [{"product": "A", "amount": 150}, {"product": "C", "amount": 300}],
    },
    {"region": "North", "sales": [{"product": "B", "amount": 250}]},
]

# Calculate total revenue per region and find the top region
# YOUR CODE HERE
```

```python
# Solution to Exercise 4
sales_data = [
    {
        "region": "North",
        "products": [
            {"name": "Widget", "quantity": 100, "price": 10},
            {"name": "Gadget", "quantity": 50, "price": 25},
        ],
    },
    {
        "region": "South",
        "products": [
            {"name": "Widget", "quantity": 80, "price": 10},
            {"name": "Gadget", "quantity": 30, "price": 25},
        ],
    },
]

# 1. Total revenue per region
print("Total revenue per region:")
for region_data in sales_data:
    region = region_data["region"]
    total = sum(p["quantity"] * p["price"] for p in region_data["products"])
    print(f"  {region}: ${total:,}")

# 2. Overall total revenue
overall_total = sum(
    p["quantity"] * p["price"]
    for region_data in sales_data
    for p in region_data["products"]
)
print(f"\nOverall total revenue: ${overall_total:,}")

# 3. Region with highest revenue
region_revenues = []
for region_data in sales_data:
    region = region_data["region"]
    total = sum(p["quantity"] * p["price"] for p in region_data["products"])
    region_revenues.append((region, total))

best_region = max(region_revenues, key=lambda x: x[1])
print(f"\nHighest revenue region: {best_region[0]} (${best_region[1]:,})")
```
